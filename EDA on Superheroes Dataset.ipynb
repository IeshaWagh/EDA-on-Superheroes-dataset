```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
marvel=pd.read_csv('characters_stats.csv')
```

```python
marvel.head()
```

```python
marvel.shape
```

```python
#to get all marvels with good alignment
good=marvel[marvel['Alignment']=='good']
```

```python
good.head()
```

```python
#sort by Intelligence
good.sort_values(by=['Intelligence'],ascending=False)[0:5]
```

```python
#sort by strength
good.sort_values(by=['Strength'],ascending=False)[0:5]
```

```python
#sort by speed
good.sort_values(by=['Speed'],ascending=False)[0:5]
```

```python
#bar graph for Alignment
plt.bar(list(marvel['Alignment'].value_counts().keys()),list(marvel['Alignment'].value_counts()))
plt.title("Bar graph for Alignment")
plt.xlabel("Alignment")
plt.ylabel("Value")
plt.show()
```

```python
#to see histogram for strength
plt.hist(list(marvel['Strength']))
plt.title("Histogram for Strength")
plt.show()
```

```python
#to see distributed plot
sns.distplot(list(marvel['Strength']))
plt.title("Distribution plot for Strength")
plt.plot()
```

```python
#to only see frequency curve
sns.distplot(list(marvel['Strength']),hist=False)
plt.plot()
```

```python
#to only see histogram
sns.distplot(list(marvel['Strength']),kde=False)
plt.show()
```

```python
#boxplot for power
plt.boxplot(marvel['Power'])
plt.title("Boxplot for Power")
plt.show()
```

```python
#records with full power
full_power=marvel[marvel['Power']==100]
```

```python
full_power.head()
```

```python
#marvel with power 100 and good alignment
pa=marvel[(marvel['Power']==100) & (marvel['Alignment']=='good')]
```

```python
pa.head()
```

```python
pa.value_counts()
```

```python
#marvel with power 100 and good alignment and combat >50
marvel[(marvel['Power']==100) & (marvel['Alignment']=='good') & (marvel['Combat']>50)]
```

```python
#marvel with power 100 and good alignment or combat>50
marvel[((marvel['Power']==100) & (marvel['Alignment']=='good') )| (marvel['Combat']>50)]
```

```python
#mean of power
marvel['Power'].mean()
```

```python
#median of combat
marvel['Combat'].median()
```

```python
#min of intelligence
marvel['Intelligence'].min()
```

```python
#max of intelligence
marvel['Intelligence'].max()
```

```python
#bar graph name vs power
plt.bar(list(marvel['Name'][0:5]),list(marvel['Power'])[0:5])
plt.title("Name v/s Power")
plt.xlabel("Name")
plt.ylabel("Power")
plt.show()
```

```python
#histogram for combat
plt.hist(marvel['Combat'],bins=7,color='r')
plt.title("Histogram for Combat")
plt.show()
```
